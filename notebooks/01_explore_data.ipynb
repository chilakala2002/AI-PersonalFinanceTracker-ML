```python
# Import required libraries
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load CSV files
transactions_df = pd.read_csv("../data/raw_data/transactions_data.csv")
users_df = pd.read_csv("../data/raw_data/users_data.csv")

# Load JSON file
with open("../data/raw_data/mcc_codes.json", "r") as file:
    mcc_data = json.load(file)

# Convert MCC JSON to DataFrame
mcc_df = pd.DataFrame(mcc_data.items(), columns=["Code", "Description"])
```

```python
# Overview of Transactions Dataset
print(" Transactions Data Overview:")
print(transactions_df.info())  # Check column names, data types, and missing values
print(transactions_df.head())  # Show first few rows

# Overview of Users Dataset
print("\n Users Data Overview:")
print(users_df.info())
print(users_df.head())

# Overview of MCC Codes Dataset
print("\n MCC Codes Overview:")
print(mcc_df.info())
print(mcc_df.head())
```

```python
# Check for missing values in Transactions Data
print("\n Missing Values in Transactions Data:")
print(transactions_df.isnull().sum())

# Check for missing values in Users Data
print("\nMissing Values in Users Data:")
print(users_df.isnull().sum())

# Check for missing values in MCC Codes
print("\nMissing Values in MCC Codes:")
print(mcc_df.isnull().sum())

# Check for duplicate values
print("\nChecking for Duplicates:")
print(f"Transactions Duplicate Rows: {transactions_df.duplicated().sum()}")
print(f"Users Duplicate Rows: {users_df.duplicated().sum()}")
print(f"MCC Duplicate Rows: {mcc_df.duplicated().sum()}")
```

```python
# Display summary statistics for numerical columns
print("\n Statistical Summary of Transactions Data:")
print(transactions_df.describe())

print("\n Statistical Summary of Users Data:")
print(users_df.describe())
```

```python
# Unique clients and cards count
print(f"\n Unique Clients: {transactions_df['client_id'].nunique()}")
print(f" Unique Cards: {transactions_df['card_id'].nunique()}")

# Unique MCC categories
print(f"\n Unique MCC Categories: {mcc_df['Description'].nunique()}")
```

```python
# Set Figure Size
plt.figure(figsize=(8, 5))

# Plot Transaction Amounts
sns.histplot(transactions_df['amount'], bins=50, kde=True, color='blue', edgecolor='black')

# This will improve Readability
plt.xlabel("Transaction Amount ($)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title("Distribution of Transaction Amounts", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show Plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure font compatibility
plt.rcParams["font.family"] = "Arial"

# Create figure
plt.figure(figsize=(8, 5))

# Corrected count plot
sns.countplot(data=users_df, x="gender", hue="gender", palette="coolwarm", legend=False)

# Add labels and title
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Distribution in Users Data")

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# defining 'top_mcc'
top_mcc = mcc_df["Description"].value_counts().head(15)

# Set default font to avoid missing glyphs
plt.rcParams["font.family"] = "DejaVu Sans"

# Create bar plot with fixes
sns.barplot(
    y=top_mcc.index,
    x=top_mcc.values,
    hue=top_mcc.index,  # Assign y-variable to hue
    palette=sns.color_palette("coolwarm", len(top_mcc)),
    legend=False
)

# Improve Readability
plt.xlabel("Frequency", fontsize=12)
plt.ylabel("MCC Category", fontsize=12)
plt.title("Top 15 Most Common MCC Categories", fontsize=14)

# Add grid for better readability
plt.grid(axis="x", linestyle="--", alpha=0.7)

# Show Plot
plt.show()
```

```python
# Function to identify categorical and numerical columns
def identify_column_types(df, dataset_name):
    categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
    numerical_cols = df.select_dtypes(include=['number']).columns.tolist()

    print(f" Dataset: {dataset_name}")
    print(f"  --> Categorical Columns: {categorical_cols}")
    print(f"  --> Numerical Columns: {numerical_cols}")
    print("-" * 80)

# Apply function to each dataset
identify_column_types(users_df, "Users Data")
identify_column_types(transactions_df, "Transactions Data")
identify_column_types(mcc_df, "MCC Codes Data")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Remove non-numeric columns from correlation analysis
users_numeric = users_df.select_dtypes(include=['number'])

# Create a correlation heatmap for numerical features
plt.figure(figsize=(12, 6))
sns.heatmap(users_numeric.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Add title for clarity
plt.title("Correlation Matrix of User Financial Data (users_data.csv)")

# Show the plot
plt.show()
```

```python
# Define numerical columns for boxplots
num_cols = ["yearly_income", "total_debt", "credit_score", "num_credit_cards", "retirement_age"]

# Create boxplots for numerical features
plt.figure(figsize=(12, 6))
users_df[num_cols].boxplot()

# Add title and labels
plt.title("Outlier Detection in User Financial Data")
plt.xticks(rotation=20)
plt.ylabel("Value")

# Show the plot
plt.show()
```
